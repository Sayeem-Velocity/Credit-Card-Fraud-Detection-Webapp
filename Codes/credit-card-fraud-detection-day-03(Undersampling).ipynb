# Dataset Preprocessing

```python
X = df.drop(columns='Class', axis=1)
y = df['Class']
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.25, random_state=42, stratify=y)
```

```python
from imblearn.under_sampling import RandomUnderSampler
import pandas as pd

undersample = RandomUnderSampler(random_state=42)
X_train_under, y_train_under = undersample.fit_resample(X_train, y_train)

print("Before Undersampling:", y_train.value_counts().to_dict())
print("After Undersampling:", pd.Series(y_train_under).value_counts().to_dict())
```

# Undersampling+GridSearchCV

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score, confusion_matrix, ConfusionMatrixDisplay, RocCurveDisplay
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

model_metrics = {}

def evaluate_model(y_true, y_pred, y_proba, model_name, model, X):
    acc = accuracy_score(y_true, y_pred)
    prec = precision_score(y_true, y_pred)
    rec = recall_score(y_true, y_pred)
    f1 = f1_score(y_true, y_pred)
    auc = roc_auc_score(y_true, y_proba)

    model_metrics[model_name] = {
        'Accuracy': acc,
        'Precision': prec,
        'Recall': rec,
        'F1-Score': f1,
        'ROC AUC': auc
    }

    print(f"\n{model_name} Metrics:")
    print(f"Accuracy: {acc:.4f}, Precision: {prec:.4f}, Recall: {rec:.4f}, F1: {f1:.4f}, AUC: {auc:.4f}")

    cm = confusion_matrix(y_true, y_pred)
    disp = ConfusionMatrixDisplay(confusion_matrix=cm)
    disp.plot(cmap='Blues')
    plt.title(f'Confusion Matrix - {model_name}')
    plt.show()

    RocCurveDisplay.from_predictions(y_true, y_proba, name=model_name)
    plt.title(f'ROC Curve - {model_name}')
    plt.show()

    if hasattr(model, 'feature_importances_'):
        plt.figure(figsize=(8, 6))
        pd.Series(model.feature_importances_, index=X.columns).sort_values().plot.barh()
        plt.title(f'Feature Importance - {model_name}')
        plt.tight_layout()
        plt.show()
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import GridSearchCV

params_dt = {
    'max_depth': [5, 10, 20, None],
    'min_samples_split': [2, 5, 10],
    'class_weight': [None, 'balanced']
}

dt = DecisionTreeClassifier(random_state=42)
grid_dt = GridSearchCV(dt, param_grid=params_dt, scoring='roc_auc', cv=3, n_jobs=-1)
grid_dt.fit(X_train_under, y_train_under)

best_dt = grid_dt.best_estimator_
y_pred = best_dt.predict(X_test)
y_proba = best_dt.predict_proba(X_test)[:, 1]

evaluate_model(y_test, y_pred, y_proba, "Decision Tree", best_dt, X_train_under)
```

```python
from sklearn.ensemble import RandomForestClassifier

params_rf = {
    'n_estimators': [100, 200],
    'max_depth': [10, 20, None],
    'min_samples_split': [2, 5],
    'class_weight': [None, 'balanced']
}

rf = RandomForestClassifier(random_state=42)
grid_rf = GridSearchCV(rf, param_grid=params_rf, scoring='roc_auc', cv=3, n_jobs=-1)
grid_rf.fit(X_train_under, y_train_under)

best_rf = grid_rf.best_estimator_
y_pred = best_rf.predict(X_test)
y_proba = best_rf.predict_proba(X_test)[:, 1]

evaluate_model(y_test, y_pred, y_proba, "Random Forest", best_rf, X_train_under)
```

```python
from xgboost import XGBClassifier

params_xgb = {
    'n_estimators': [100, 200],
    'max_depth': [3, 5, 10],
    'learning_rate': [0.01, 0.1],
    'scale_pos_weight': [1, 10]
}

xgb = XGBClassifier(random_state=42, use_label_encoder=False, eval_metric='logloss')
grid_xgb = GridSearchCV(xgb, param_grid=params_xgb, scoring='roc_auc', cv=3, n_jobs=-1)
grid_xgb.fit(X_train_under, y_train_under)

best_xgb = grid_xgb.best_estimator_
y_pred = best_xgb.predict(X_test)
y_proba = best_xgb.predict_proba(X_test)[:, 1]

evaluate_model(y_test, y_pred, y_proba, "XGBoost", best_xgb, X_train_under)
```

```python
from lightgbm import LGBMClassifier

params_lgbm = {
    'n_estimators': [100, 200],
    'max_depth': [5, 10, -1],
    'learning_rate': [0.01, 0.1],
    'class_weight': [None, 'balanced']
}

lgbm = LGBMClassifier(random_state=42)
grid_lgbm = GridSearchCV(lgbm, param_grid=params_lgbm, scoring='roc_auc', cv=3, n_jobs=-1)
grid_lgbm.fit(X_train_under, y_train_under)

best_lgbm = grid_lgbm.best_estimator_
y_pred = best_lgbm.predict(X_test)
y_proba = best_lgbm.predict_proba(X_test)[:, 1]

evaluate_model(y_test, y_pred, y_proba, "LightGBM", best_lgbm, X_train_under)
```

```python
from sklearn.metrics import roc_curve, auc
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 8))

for model_name, model in {
    "Decision Tree": best_dt,
    "Random Forest": best_rf,
    "XGBoost": best_xgb,
    "LightGBM": best_lgbm
}.items():
    y_proba = model.predict_proba(X_test)[:, 1]
    fpr, tpr, _ = roc_curve(y_test, y_proba)
    roc_auc = auc(fpr, tpr)
    plt.plot(fpr, tpr, label=f"{model_name} (AUC = {roc_auc:.4f})")

plt.plot([0, 1], [0, 1], 'k--', label="Random Guess")
plt.title("Combined ROC Curve")
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.grid(True)
plt.legend(loc="lower right")
plt.tight_layout()
plt.show()
```

### 🔍 **Model Performance Analysis & Insights**

We evaluated the performance of four classification algorithms—**Decision Tree, Random Forest, XGBoost,** and **LightGBM**—using various performance metrics, including Accuracy, Precision, Recall, F1-Score, and ROC AUC. Below are the key insights:

#### 📊 **Performance Metrics Overview**

| Model             | Accuracy   | Precision | Recall    | F1-Score  | ROC AUC   |
| ----------------- | ---------- | --------- | --------- | --------- | --------- |
| **XGBoost**       | 93.21%     | 0.023     | **0.934** | 0.045     | **0.980** |
| **LightGBM**      | 95.47%     | 0.034     | 0.911     | 0.065     | 0.978     |
| **Random Forest** | **97.48%** | **0.058** | 0.894     | **0.109** | 0.977     |
| **Decision Tree** | 90.31%     | 0.016     | 0.911     | 0.031     | 0.936     |

#### ✅ **Key Takeaways**

1. **ROC AUC**:

   * All models perform well above the baseline, with **XGBoost (0.9802)** slightly outperforming others, indicating excellent discriminatory ability.
   * **LightGBM** and **Random Forest** closely follow, with ROC AUCs of 0.9772 and 0.9766, respectively.

2. **Accuracy & F1-Score**:

   * **Random Forest** achieved the **highest accuracy (97.48%)** and the **highest F1-Score (0.109)**, suggesting a good balance between precision and recall.
   * **LightGBM** follows with high accuracy (95.47%) and a respectable F1-Score (0.065).

3. **Precision vs Recall**:

   * All models show relatively **low precision** but **high recall**, which often indicates class imbalance where positive cases are rare.
   * **XGBoost** has the highest recall (0.935), making it well-suited for applications where minimizing false negatives is critical (e.g., medical diagnosis or fraud detection).

#### ⚠️ **Possible Considerations**

* The **low precision** across all models signals a **high false-positive rate**, which might be undesirable in some applications.
* The **class imbalance** should be investigated and possibly mitigated (e.g., through SMOTE, class weighting, or threshold tuning).
