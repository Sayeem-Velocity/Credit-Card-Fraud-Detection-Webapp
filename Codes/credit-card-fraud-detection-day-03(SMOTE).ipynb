# Dataset Preprocessing

```python
X = df.drop(columns='Class', axis=1)
y = df['Class']
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42, stratify=y)
```

```python
from imblearn.over_sampling import SMOTE
import pandas as pd

smote = SMOTE(random_state=42)
X_train_smote, y_train_smote = smote.fit_resample(X_train, y_train)

print("Before SMOTE:", y_train.value_counts().to_dict())
print("After SMOTE:", pd.Series(y_train_smote).value_counts().to_dict())
```

# Hypertuned Models

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score, confusion_matrix, ConfusionMatrixDisplay, RocCurveDisplay
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

model_metrics = {}

def evaluate_model(y_true, y_pred, y_proba, model_name, model, X):
    acc = accuracy_score(y_true, y_pred)
    prec = precision_score(y_true, y_pred)
    rec = recall_score(y_true, y_pred)
    f1 = f1_score(y_true, y_pred)
    auc = roc_auc_score(y_true, y_proba)

    model_metrics[model_name] = {
        'Accuracy': acc,
        'Precision': prec,
        'Recall': rec,
        'F1-Score': f1,
        'ROC AUC': auc
    }

    print(f"\n{model_name} Metrics:")
    print(f"Accuracy: {acc:.4f}, Precision: {prec:.4f}, Recall: {rec:.4f}, F1: {f1:.4f}, AUC: {auc:.4f}")

    # Confusion Matrix
    cm = confusion_matrix(y_true, y_pred)
    disp = ConfusionMatrixDisplay(confusion_matrix=cm)
    disp.plot(cmap='Blues')
    plt.title(f'Confusion Matrix - {model_name}')
    plt.show()

    # ROC Curve
    RocCurveDisplay.from_predictions(y_true, y_proba, name=model_name)
    plt.title(f'ROC Curve - {model_name}')
    plt.show()

    # Feature Importance
    if hasattr(model, 'feature_importances_'):
        plt.figure(figsize=(8, 6))
        pd.Series(model.feature_importances_, index=X.columns).sort_values().plot.barh()
        plt.title(f'Feature Importance - {model_name}')
        plt.tight_layout()
        plt.show()
```

# Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import GridSearchCV

params_dt = {
    'max_depth': [3, 5, 10],
    'min_samples_split': [2, 5],
    'class_weight': ['balanced']
}

grid_dt = GridSearchCV(DecisionTreeClassifier(random_state=42),
                       param_grid=params_dt, scoring='roc_auc', cv=3, n_jobs=-1)
grid_dt.fit(X_train_smote, y_train_smote)
best_dt = grid_dt.best_estimator_

y_pred = best_dt.predict(X_test)
y_proba = best_dt.predict_proba(X_test)[:, 1]

evaluate_model(y_test, y_pred, y_proba, "Decision Tree", best_dt, X_train_smote)
```

# Random Forest

```python
from sklearn.ensemble import RandomForestClassifier

params_rf = {
    'n_estimators': [100],
    'max_depth': [10, None],
    'min_samples_split': [2],
    'class_weight': ['balanced']
}

grid_rf = GridSearchCV(RandomForestClassifier(random_state=42),
                       param_grid=params_rf, scoring='roc_auc', cv=3, n_jobs=-1)
grid_rf.fit(X_train_smote, y_train_smote)
best_rf = grid_rf.best_estimator_

y_pred = best_rf.predict(X_test)
y_proba = best_rf.predict_proba(X_test)[:, 1]

evaluate_model(y_test, y_pred, y_proba, "Random Forest", best_rf, X_train_smote)
```

# Gradient Boosting

```python
from xgboost import XGBClassifier

params_xgb = {
    'n_estimators': [100],
    'max_depth': [3, 6],
    'learning_rate': [0.1],
    'scale_pos_weight': [1, sum(y_train_smote == 0) / sum(y_train_smote == 1)]
}

grid_xgb = GridSearchCV(XGBClassifier(random_state=42, use_label_encoder=False, eval_metric='logloss'),
                        param_grid=params_xgb, scoring='roc_auc', cv=3, n_jobs=-1)
grid_xgb.fit(X_train_smote, y_train_smote)
best_xgb = grid_xgb.best_estimator_

y_pred = best_xgb.predict(X_test)
y_proba = best_xgb.predict_proba(X_test)[:, 1]

evaluate_model(y_test, y_pred, y_proba, "XGBoost", best_xgb, X_train_smote)
```

# LightGBM

```python
from lightgbm import LGBMClassifier

params_lgbm = {
    'n_estimators': [100],
    'max_depth': [5, -1],
    'learning_rate': [0.1],
    'class_weight': ['balanced']
}

grid_lgbm = GridSearchCV(LGBMClassifier(random_state=42),
                         param_grid=params_lgbm, scoring='roc_auc', cv=3, n_jobs=-1)
grid_lgbm.fit(X_train_smote, y_train_smote)
best_lgbm = grid_lgbm.best_estimator_

y_pred = best_lgbm.predict(X_test)
y_proba = best_lgbm.predict_proba(X_test)[:, 1]

evaluate_model(y_test, y_pred, y_proba, "LightGBM", best_lgbm, X_train_smote)
```

# Model Comparison

```python
from sklearn.metrics import roc_curve, auc
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 8))

for model_name, model in {
    "Decision Tree": best_dt,
    "Random Forest": best_rf,
    "XGBoost": best_xgb,
    "LightGBM": best_lgbm
}.items():
    y_proba = model.predict_proba(X_test)[:, 1]
    fpr, tpr, _ = roc_curve(y_test, y_proba)
    roc_auc = auc(fpr, tpr)
    plt.plot(fpr, tpr, label=f"{model_name} (AUC = {roc_auc:.4f})")

# Add diagonal line for random guess
plt.plot([0, 1], [0, 1], 'k--', label="Random Guess")

plt.title("Combined ROC Curve")
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.grid(True)
plt.legend(loc="lower right")
plt.tight_layout()
plt.show()
```

# Insights

### 📊 **Model Performance Analysis**

We evaluated four classification models—**Decision Tree, Random Forest, XGBoost, and LightGBM**—using multiple performance metrics: Accuracy, Precision, Recall, F1-Score, and ROC AUC. The results demonstrate the effectiveness of ensemble methods, especially **XGBoost** and **LightGBM**, in this classification task.

---

### 🔍 **Key Observations**

#### ✅ **1. ROC Curve & AUC Scores**

* The ROC curve clearly shows that **XGBoost (AUC = 0.9726)** and **LightGBM (AUC = 0.9682)** outperform other models in terms of overall discriminative ability.
* **Random Forest** also performs well with an AUC of **0.9493**, indicating strong classification capability.
* **Decision Tree**, while simple and interpretable, has the lowest AUC (**0.8825**), confirming its limitations in capturing complex decision boundaries.

#### 📈 **2. Accuracy**

* All models achieved very high accuracy (>99%), but **XGBoost** leads with **99.89%**, followed closely by **Random Forest** (**99.94%**) and **LightGBM** (**99.82%**).
* The high accuracy across all models may be influenced by class imbalance or dominant class presence.

#### 📊 **3. Precision vs. Recall**

* **Decision Tree** shows a **very low precision (0.1571)** but high recall (0.7973), indicating a high number of false positives.
* **Random Forest** balances both precision (**0.8741**) and recall (**0.7973**) well, resulting in the best **F1-Score (0.8339)** among all models.
* **XGBoost** and **LightGBM** slightly sacrifice precision for higher recall:

  * XGBoost: Precision = 0.6294, Recall = 0.8378
  * LightGBM: Precision = 0.4808, Recall = 0.8446
* This tradeoff might be acceptable in domains where missing positives is more critical than false alarms (e.g., medical diagnostics).

---

### 🏆 **Best Performing Model**

Based on the **balanced trade-off** between **precision, recall, and AUC**, **Random Forest** emerges as the most **robust model** in this setting. However, if **recall is the priority**, **XGBoost** or **LightGBM** may be preferable due to their slightly higher recall and AUC scores.
